#### 0) Imports

```python
from __future__ import absolute_import, division, print_function
from __future__ import unicode_literals

import tensorflow        as tf
import numpy             as np
import matplotlib.pyplot as plt

from tensorflow              import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
print(tf.__version__)
```

# 1) télécharger le data set

```python
fashion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

# 2) Caractèristiques du problème

```python
nb_de_classe_a_predire = 10
largeur_image = 28
hauteur_image = 28
```

# 3) Normalisation des valeurs 

```python
train_images[0]
```

```python
train_images_normalized = train_images / 255.0
test_images_normalized  = test_images / 255.0

# Does CNN work more efficiently on the range [0…1]?
# Yes. It is easier for the neural network to find a local or a global mininum (either one is fine) on a smaller range of 0-1 than 0-255.
# from https://www.kaggle.com/c/digit-recognizer/discussion/102458
```

```python
if True:
    plt.figure(figsize=(10,10))
    for i in range(25):
        plt.subplot(5,5,i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.imshow(train_images[i])#, cmap=plt.cm.binary)
        plt.xlabel(class_names[train_labels[i]])
    plt.show()
```

# 4) définir les couches du réseau de neurone

## 4.1) couche 1->  flatten = "transformer la matrice 60 000 * 28 * 28 en une matrice 60 000 * 784"

```python
input_layer_flattener = keras.layers.Flatten(
                          input_shape=( largeur_image, 
                                        hauteur_image))

flattened = input_layer_flattener.apply(train_images)
flattened.get_shape(), train_images.shape
```

```python
flattened[0]
```

## 4.2) couche 2 -> couche cachée

```python
nb_de_neurone       = nb_de_classe_a_predire *2
activation_fonction = "relu"
couche_cachee_relu  = Dense( nb_de_neurone,
                             activation  = activation_fonction)
```

## 4.3)  Couche 3 -> couche de sortie

```python
nb_de_neurone_en_sortie = nb_de_classe_a_predire
fonction                = "softmax"
couche_de_sortie        = Dense(nb_de_neurone_en_sortie,
                                activation=fonction)
```

# 5) Créer un modèle
## 5.1) Créer le réseau de neurones

```python
modele = keras.Sequential()
modele.add(input_layer_flattener)
modele.add(couche_cachee_relu)
modele.add(couche_de_sortie)
```

## 5.2) Compiler le modèle

```python
modele.compile(optimizer = 'adam',
               loss      = 'sparse_categorical_crossentropy',
               metrics   = ['accuracy'])
modele.summary()
```

## 5.3) Apprendre

```python
history_100_epochs = modele.fit(train_images_normalized, 
           train_labels, 
           epochs=100,
           batch_size=500)
```

```python
history_2_epochs = modele.fit(train_images_normalized, 
           train_labels, 
           epochs=2,
           batch_size=32)
```

# 6) Analyser les résultats

## 6.1)  Faire des prédictions

```python
predictions = modele.predict(test_images)
predictions[0]
```

```python
classe = np.argmax(predictions[0])
print(f"classe = {classe}, soit [{class_names[classe]}]")
_ = plt.imshow(test_images[0])
```

## 6.2) Calculer la loss

```python
test_loss, test_acc = modele.evaluate(test_images,  
                                      test_labels, 
                                      verbose=2)

print('\nTest accuracy: %.2f'% test_acc)
```

## 6.3) afficher toutes les prédictions et images 

```python
def plot_image(i, predictions_array, true_label, img):
    predictions_array, true_label, img = predictions_array, true_label[i], img[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])

    plt.imshow(img, cmap=plt.cm.binary)

    predicted_label = np.argmax(predictions_array)
    if predicted_label == true_label:
        color = 'blue'
    else:
        color = 'red'

    plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label],
                                100*np.max(predictions_array),
                                class_names[true_label]),
                                color=color)

def plot_value_array(i, predictions_array, true_label):
    predictions_array, true_label = predictions_array, true_label[i]
    plt.grid(False)
    plt.xticks(range(10))
    plt.yticks([])
    thisplot = plt.bar(range(10), predictions_array, color="#777777")
    plt.ylim([0, 1])
    predicted_label = np.argmax(predictions_array)

    thisplot[predicted_label].set_color('red')
    thisplot[true_label].set_color('blue')
```

```python

i = 0
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plot_image(i, predictions[i], test_labels, test_images)
plt.subplot(1,2,2)
plot_value_array(i, predictions[i],  test_labels)
plt.show()
```

```python
# Plot the first X test images, their predicted labels, and the true labels.
# Color correct predictions in blue and incorrect predictions in red.
num_rows = 5
num_cols = 3
num_images = num_rows*num_cols
plt.figure(figsize=(2*2*num_cols, 2*num_rows))
for i in range(num_images):
  plt.subplot(num_rows, 2*num_cols, 2*i+1)
  plot_image(i, predictions[i], test_labels, test_images)
  plt.subplot(num_rows, 2*num_cols, 2*i+2)
  plot_value_array(i, predictions[i], test_labels)
plt.tight_layout()
plt.show()
```
